```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
y = data['Delivery_person_Ratings']
y
```

```python
x = data.drop(['Delivery_person_Ratings'], axis = 1)
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 100)
```

```python
model = DecisionTreeRegressor()
```

```python
model_prep = model.fit(x_train,y_train)
```

```python
y_predict= model_prep.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE = mean_squared_error(y_test, y_predict)
print("Mean Squared Error:", MSE)
```

```python
import numpy as np
```

```python
rmse = np.sqrt(MSE)
print("Root Mean Squared Error:", rmse)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, y_predict)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_predict)
print("R-squared:", r2)
```

we'll be using different models to check wheather they fit here better or not.

##### Using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=50)
model.fit(x_train, y_train)
```

```python
model_prediction = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, model_prediction)
r2 = r2_score(y_test, model_prediction)

print("Mean Squared Error:", mse)
print("R-squared:", r2) 
```

##### Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
regr.fit(x_train,y_train)
Y_pred = regr.predict(x_test)
```

```python
MEAN = mean_squared_error(y_test,Y_pred)
R2 = r2_score(y_test,Y_pred)

print("Mean Squared Error:", MEAN)
print("R-squared:", R2)
```

### Conclusion:

It can be seen that Linear Regression is a better prediction model comparing to Decision Tree or Random Forest as the showed R2 being negative whereas R2 shifted to positive (i.e. 0.029)in Linear Regression telling that Linear Regression is a better predictor.

Though this is a room for improvement and various other models can be tested to find a better accuracy.
